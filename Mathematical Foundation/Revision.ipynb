## 1. Xác suất và Thống kê (Probabilty & Statistics)

* **Xác suất có điều kiện & Định lý Bayes**:
  * Cần nắm vững công thức Bayes để tính xác suất ngược:
  $$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$
  * Phân biệt giữa các biến cố độc lập và biến cố xung khắc

* **Ước lượng hợp lý cực đại - MLE**
  * Cách thiết lập hàm Likelihood $L(\lambda)$ và tìm giá trị $\lambda$ làm cực đại hàm đó
  * Ước lượng tỷ lệ ($p$) trong phân phối Bernoulli.
  * **Phân phối mũ:** Nhớ các đặc trưng cơ bản như $E[X] = 1/\lambda$ và $Var(X) = 1/\lambda^2$
  
* **Kiểm định giả thuyết**
  * Cách tính giá trị kiểm định $z = \frac{\overline{x} - \mu}{\sigma / \sqrt{n}}$
  * Quy tắc bác bỏ $H_0$:
  1. **Sử dụng Giá trị P (P-value):** Đây là phương pháp hiện đại và phổ biến nhất, đặc biệt là khi sử dụng các phần mềm thống kê.
      * **P-value:** Là xác suất tìm thấy kết quả cực đoan như kết quả quan sát được (hoặc hơn thế) nếu giả thuyết $H_0$ là đúng.
      * **Mức ý nghĩa ($\alpha$):** Thường được chọn trước là 0.05 (5%), 0.01 (1%), hoặc 0.10 (10%).
      
      **Quy tắc quyết định:**
      
      * Nếu $P\text{-value} \le \alpha$: Bác bỏ $H_0$. Kết quả có ý nghĩa thống kê.
      * Nếu $P\text{-value} > \alpha$: Không đủ cơ sở để bác bỏ $H_0$ (Chấp nhận $H_0$).
      
      **Mẹo nhớ** "If the P is low, the $H_0$ must go" (Nếu P thấp, $H_0$ phải "đi").
      
  2. **Sử dụng Giá trị Tới hạn (Critical Value):** Phương pháp này dựa trên việc so sánh giá trị thống kê tính toán được từ mẫu với một ngưỡng cắt trên phân phối xác suất.
      * **Miền bác bỏ ($W$):** Là tập hợp các giá trị dẫn đến việc bác bỏ $H_0$.
      * **Giá trị kiểm định ($Z_{qs}$ hoặc $T_{qs}$):** Giá trị tính toán được từ dữ liệu thực tế.
      * **Giá trị tới hạn ($Z_{\alpha}$ hoặc $t_{\alpha}$):** Giá trị tra bảng dựa trên $\alpha$.
      
      **Quy tắc quyết định:**
        * **Kiểm định hai phía ($H_1: \mu \neq \mu_0$):** Bác bỏ $H_0$ nếu $|Z_{qs}| > Z_{\alpha/2}$.
        * **Kiểm định phía phải ($H_1: \mu > \mu_0$):** Bác bỏ $H_0$ nếu $Z_{qs} > Z_{\alpha}$.
        * **Kiểm định phía trái ($H_1: \mu < \mu_0$):** Bác bỏ $H_0$ nếu $Z_{qs} < -Z_{\alpha}$

* **Hiệp phương sai - Covariance**
  * Công thức $Cov(X, Y) = E[XY] - E[X]E[Y]$
  * Nếu $X, Y$ độc lập và tuân theo $\mathcal{N}(0,1)$, cần nhớ các tính chất của moment bậc cao để tính $Cov(Z^2, W^2)$
  * Tính độ lệch chuẩn của hiệu hai biến ngẫu nhiên độc lập $X - Y$: $\sigma_{X-Y} = \sqrt{\sigma_X^2 + \sigma_Y^2}$.

## 2. Lý thuyết Đồ thị và Cây (Graph Theory & Trees)

* **Cấu trúc Cây**
  * **Cây $m$-phân đầy đủ:** Mối liên hệ giữa số nút lá ($L$), số nút trong ($I$) và tổng số nút ($n$):
    * $n = m \cdot I + 1$
    * $L = (m-1)I + 1 \Rightarrow I = \frac{L-1}{m-1}$

  * **Cây nhị phân (Binary Tree):** Mối quan hệ giữa số lượng nút lá ($L$) và số lượng nút có đúng 2 con ($N_2$). Công thức: $N_2 = L - 1$.

* **Đặc điểm Đồ thị phẳng**
  * Công thức Euler: $v - e + f = 2$ (với $v$: đỉnh, $e$: cạnh, $f$: mặt)
  * Đối với đồ thị tam giác hóa (mỗi mặt bao bởi 3 cạnh): $3f = 2e$
  * Số sắc số ($\chi(G)$) và các định lý liên quan đến đồ thị chứa tam giác.
  * Đồ thị Hamiltonian và tính phẳng.
  
* **Các khái niệm nâng cao**
  * **Crossing number:** Số giao điểm tối thiểu khi vẽ đồ thị trên mặt phẳng13.
  * **Thành phần liên thông:** Rừng (forest) gồm $k$ cây với $n$ đỉnh và $e$ cạnh có $k = n - e$ thành phần. Thêm 1 cạnh không tạo chu trình sẽ làm giảm số thành phần liên thông xuống 1 ($k-1$)
  * **Đường đi ngắn nhất:** Hiểu các tính chất về đường đi đơn và sự tồn tại của đỉnh trung gian trên mọi đường đi ngắn nhất.
  * **Cộng đồng trong đồ thị (Community Detection):** Khái niệm về độ đo Modularity ($Q$) và cách tính toán dựa trên ma trận kề và bậc của các đỉnh.

## 3. Đại số tuyến tính và Tối ưu hóa (Linear Algebra & Optimization)

* **Ma trận nghịch đảo và Chuyển vị**
  * Tính chất: $(AB)^T = B^T A^T$ và $(A^{-1})^T = (A^T)^{-1}$
  * Giải phương trình ma trận dạng $(A^{-1}BC^T)^T = I$ để cô lập $B$

* **Cực trị có điều kiện (Lagrange Multipliers):** Cách tìm giá trị lớn nhất/nhỏ nhất của một hàm số (ví dụ: $(xy)^{2024}$) trên một miền ràng buộc (ví dụ: hình ellipse $\frac{x^2}{2024} + y^2 = 1$).

* **Giải tích đa biến & Điểm dừng**
  * Tìm điểm dừng bằng cách giải hệ $\nabla f(x, y) = 0$
  * **Phân loại điểm dừng:** Sử dụng ma trận Hessian ($D = f_{xx}f_{yy} - f_{xy}^2$) để xác định cực đại, cực tiểu hay điểm yên ngựa

* **Gradient Descent**
  * Công thức cập nhật: $x_{k+1} = x_k - \eta \nabla f(x_k)$
  * Cách tính Step size ($\eta$) để đạt được mục tiêu sau một bước lặp.
  * **Hàm mất mát (Loss Function):** Cách tính Gradient cho các hàm có dạng Regularization (L2 norm) như $\nabla_w ||Xw-Y||_2^2 = 2X^T(Xw-Y)$.

* **Trị riêng & Vectơ riêng (Eigenvalues & Eigenvectors):**
  * Xác định trị riêng tương ứng với một vectơ riêng cho trước bằng phương trình $Av = \lambda v$.
  * Tìm không gian vectơ riêng (Eigenvector space) tương ứng với một trị riêng $\lambda$ cụ thể.

* **Logic vị từ và mệnh đề:**
  * **Vị từ:** Cách biểu diễn ngôn ngữ tự nhiên sang ký hiệu logic $\forall$ (với mọi) và $\exists$ (tồn tại).
  * **Tính thỏa được (Satisfiability):** Mệnh đề thỏa được nếu có ít nhất một cách gán giá trị làm mệnh đề đúng.
  * **Suy luận logic:** Giải quyết các bài toán "thủ phạm" dựa trên giả thiết về số lượng người nói thật/nói dối.


* **Lý thuyết thông tin:**
  * **Khoảng cách KL (Kullback-Leibler Divergence):** $KL(P||Q) = \sum P(x) \log \frac{P(x)}{Q(x)}$
