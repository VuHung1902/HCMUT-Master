## 1. Xác suất và Thống kê (Probabilty & Statistics)

* **Xác suất có điều kiện & Định lý Bayes**:
  * Cần nắm vững công thức Bayes để tính xác suất ngược:
  $$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$
  * Phân biệt giữa các biến cố độc lập và biến cố xung khắc
* **Ước lượng hợp lý cực đại - MLE**
  * Cách thiết lập hàm Likelihood $L(\lambda)$ và tìm giá trị $\lambda$ làm cực đại hàm đó
  * **Phân phối mũ:** Nhớ các đặc trưng cơ bản như $E[X] = 1/\lambda$ và $Var(X) = 1/\lambda^2$
  
* **Kiểm định giả thuyết**
  * Cách tính giá trị kiểm định $z = \frac{\overline{x} - \mu}{\sigma / \sqrt{n}}$
  * Quy tắc bác bỏ $H_0$: Nếu $|z| > [cite_start]z_{\alpha/2}$ thì bác bỏ giả thuyết không

* **Hiệp phương sai - Covariance**
  * Công thức $Cov(X, Y) = E[XY] - E[X]E[Y]$
  * Nếu $X, Y$ độc lập và tuân theo $\mathcal{N}(0,1)$, cần nhớ các tính chất của moment bậc cao để tính $Cov(Z^2, W^2)$9

## 2. Lý thuyết Đồ thị và Cây (Graph Theory & Trees)

* **Cấu trúc Cây**
  * **Cây $m$-phân đầy đủ:** Mối liên hệ giữa số nút lá ($L$), số nút trong ($I$) và tổng số nút ($n$):
    * $n = m \cdot I + 1$
    * $L = (m-1)I + 1 \Rightarrow I = \frac{L-1}{m-1}$
* **Đặc điểm Đồ thị phẳng**
  * Công thức Euler: $v - e + f = 2$ (với $v$: đỉnh, $e$: cạnh, $f$: mặt)
  * Đối với đồ thị tam giác hóa (mỗi mặt bao bởi 3 cạnh): $3f = 2e$
  
* **Các khái niệm nâng cao**
  * **Crossing number:** Số giao điểm tối thiểu khi vẽ đồ thị trên mặt phẳng13.
  * **Thành phần liên thông:** Rừng (forest) gồm $k$ cây với $n$ đỉnh và $e$ cạnh có $k = n - e$ thành phần14141414. Thêm 1 cạnh không tạo chu trình sẽ làm giảm số thành phần liên thông xuống 1 ($k-1$)
  * **Đường đi ngắn nhất:** Hiểu các tính chất về đường đi đơn và sự tồn tại của đỉnh trung gian trên mọi đường đi ngắn nhất.

## 3. Đại số tuyến tính và Tối ưu hóa (Linear Algebra & Optimization)

* **Ma trận nghịch đảo và Chuyển vị**
  * Tính chất: $(AB)^T = B^T A^T$ và $(A^{-1})^T = (A^T)^{-1}$
  * Giải phương trình ma trận dạng $(A^{-1}BC^T)^T = I$ để cô lập $B$
* **Giải tích đa biến & Điểm dừng**
  * Tìm điểm dừng bằng cách giải hệ $\nabla f(x, y) = 0$
  * **Phân loại điểm dừng:** Sử dụng ma trận Hessian ($D = f_{xx}f_{yy} - f_{xy}^2$) để xác định cực đại, cực tiểu hay điểm yên ngựa
* **Gradient Descent**
  * Công thức cập nhật: $x_{k+1} = x_k - \eta \nabla f(x_k)$
  * **Hàm mất mát (Loss Function):** Cách tính Gradient cho các hàm có dạng Regularization (L2 norm) như $\nabla_w ||Xw-Y||_2^2 = 2X^T(Xw-Y)$.
