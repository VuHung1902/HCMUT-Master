# Tổng hợp Bài toán Tin học Lý thuyết và Thuật toán

## Phần I: Quy hoạch Động và Thuật toán Đồ thị

### 1. Bài toán: Dãy con tăng có tổng lớn nhất (Maximum Sum Increasing Subsequence - MSIS)

* **Dãy số:** $A = [9, 1, 10, 6, 7, 3, 5, 0, 8, 4, 2, 11]$
* **Phương pháp:** Quy hoạch động (Dynamic Programming). Tính $DP[i]$ là tổng lớn nhất của dãy con tăng kết thúc tại $A[i]$.
    $$DP[i] = A[i] + \max(\{DP[j] \mid j < i \text{ và } A[j] < A[i]\} \cup \{0\})$$
* **Kết quả:** $\max(DP) = \mathbf{33}$ (từ dãy con $[1, 6, 7, 8, 11]$).
* **Kiến thức liên quan:** Bài toán tối ưu (Optimization Problem), Quy hoạch động.

### 2. Bài toán: Lợi nhuận tối đa 2 giao dịch trên thị trường chứng khoán

* **Dãy giá:** $P = [10, 22, 5, 75, 65, 80]$ USD.
* **Phương pháp:** Quy hoạch động. Chia bài toán thành 2 giai đoạn (giao dịch 1 kết thúc trước hoặc tại ngày $i$, giao dịch 2 bắt đầu từ ngày $i$).
    $$\text{Max Profit} = \max_{i} (\text{Profit}_1[i] + \text{Profit}_2[i])$$
* **Kết quả:** Lợi nhuận tối đa là $\mathbf{87}$ USD (Giao dịch 1: $22-10=12$; Giao dịch 2: $80-5=75$).
* **Kiến thức liên quan:** Bài toán tối ưu, Quy hoạch động trên mảng, Lợi nhuận tối đa (Buy/Sell Stock II).

### 3. Bài toán: Thuật toán Karger (Xác suất tìm Min-Cut)

* **Đồ thị:** $G$ có $n=6$ đỉnh. Kích thước Min-Cut $k=1$.
* **Công thức:** Xác suất tối thiểu để tìm được Min-Cut sau một lần chạy thuật toán Karger là $P \ge \frac{2}{n(n-1)}$.
    $$P \ge \frac{2}{6(5)} = \frac{1}{15} \approx 0.0667$$
* **Kết quả (Trắc nghiệm):** Chọn $\mathbf{0.27}$ (Đáp án trắc nghiệm chấp nhận).
* **Kiến thức liên quan:** Thuật toán ngẫu nhiên hóa (Randomized Algorithm), Lát cắt cực tiểu (Min-Cut), Lý thuyết đồ thị.

## Phần II: Lý thuyết Độ phức tạp (Complexity Theory)



### 4. Bài toán: Phân loại P và Q (Thu giảm về 3-SAT)

* **Dữ kiện:** $P \le_p 3-SAT$ và $3-SAT \le_p Q$.
* **Phân tích:**
    * $P \le_p 3-SAT \implies P \in \mathbf{NP}$ (Vì $3-SAT \in NP$).
    * $3-SAT \le_p Q \implies Q \in \mathbf{NP-hard}$ (Vì $3-SAT \in NP-Hard$).
* **Kết quả:** $P \in NP$, $Q \in NP-hard$.
* **Kiến thức liên quan:** Lớp độ phức tạp $NP$, $NP-Hard$, Thu giảm đa thức ($\le_p$).

### 5. Bài toán: Phân loại Q và R (S là NPC)

* **Dữ kiện:** $S$ là **NP-đầy đủ (NPC)**; $Q \le_p S$; $S \le_p R$.
* **Phân tích:**
    * $S \le_p R \implies R \in \mathbf{NP-hard}$ (Vì $S \in NP-Hard$).
    * $Q \le_p S \implies Q \in \mathbf{NP}$ (Vì $S \in NP$).
* **Kết quả:** $R \in NP-hard$.
* **Kiến thức liên quan:** $NP-Complete$, $NP-Hard$, Tính chất của thu giảm.

### 6. Bài toán: Quan hệ thu giảm không đúng

* **Khẳng định không đúng:** $\mathbf{3-SAT \le_p 2-SAT}$.
* **Lý do:** $3-SAT$ là $NPC$ (khó), $2-SAT$ thuộc $P$ (dễ). Bài toán khó không thể thu giảm về bài toán dễ trong thời gian đa thức (trừ khi $P = NP$).
* **Kiến thức liên quan:** Quan hệ thứ tự giữa các lớp $P$ và $NPC$.

### 7. Bài toán: Bài toán sử dụng tiếp cận bao phủ đỉnh (Vertex Cover)

* **Bài toán được giải bằng cách sử dụng/thu giảm về Vertex Cover:** $\mathbf{Traveling\; Salesman}$.
* **Kiến thức liên quan:** Bài toán $NPC$, Vertex Cover, Thuật toán xấp xỉ (Approximation Algorithms).

## Phần III: Xác suất và Thuật toán Ngẫu nhiên

### 8. Bài toán: Số lần lặp tối thiểu (Monte Carlo)

* **Yêu cầu:** Tìm số lần lặp $n$ để xác suất có ít nhất một kết quả đúng $\ge p_2$ (với $P(\text{đúng}) = p_1$ trong một lần lặp).
* **Công thức:** Dẫn đến bất đẳng thức $n \ge \frac{\log(1 - p_2)}{\log(1 - p_1)}$.
* **Kết quả:** $\mathbf{B. \frac{\log(1 - p_2)}{\log(1 - p_1)}}$.
* **Kiến thức liên quan:** Thuật toán Monte Carlo, Bất đẳng thức xác suất, Logarit.

### 9. Bài toán: Xác suất tìm thấy sau đúng $k$ lần

* **Yêu cầu:** Xác suất tìm thấy $a$ sau **đúng $k$ lần lặp**.
* **Công thức:** $P = (\text{thất bại})^{k-1} \times (\text{thành công})$. Giả định $P(\text{thành công}) = 1/2$.
    $$P = \left(\frac{1}{2}\right)^{k-1} \times \left(\frac{1}{2}\right) = \left(\frac{1}{2}\right)^k$$
* **Kết quả:** $\mathbf{A. (1/2)^k}$.
* **Kiến thức liên quan:** Phân phối hình học (Geometric distribution), Xác suất độc lập.

### 15. Bài toán: Công thức truy hồi lát gạch

* **Yêu cầu:** Lát gạch $1 \times 2$ và $2 \times 2$ lên sàn $2 \times N$.
* **Công thức:** Xét cách lát cột cuối cùng:
    $$f(N) = f(N-1) + f(N-2) + f(N-2) = f(N-1) + 2 \times f(N-2)$$
* **Kiến thức liên quan:** Bài toán đếm (Counting Problem), Quy hoạch động, Công thức truy hồi.

### 16. Bài toán: Giới hạn cần thiết cho Knapsack bằng Quy hoạch động

* **Kiến thức liên quan:** Bài toán Knapsack 0/1, Phân loại P và NP, Thuật toán giả đa thức (Pseudo-polynomial time).

## Phần VI: Độ phức tạp Thời gian và Phân tích Thuật toán

### 17. Bài toán: Độ phức tạp thời gian của đoạn mã

* **Đoạn mã:** 3 vòng lặp lồng nhau: $i=1..n, j=1..n, k=1..10$.
* **Tính toán:** $T(n) = n \times n \times 10 = 10n^2$.
* **Độ phức tạp:** $\mathbf{O(n^2)}$.
* **Kiến thức liên quan:** Ký hiệu Big O, Phân tích thuật toán.

### 18. Bài toán: Độ phức tạp của Thuật toán Đệ quy

* **Phương trình truy hồi:** $T(n) = 2T\left(\frac{n}{3}\right) + \Theta(n^2)$.
* **Phương pháp:** Định lý Thợ (Master Theorem): $a=2, b=3, f(n)=n^2$.
    * So sánh $f(n)$ với $n^{\log_3 2} \approx n^{0.63}$.
    * $f(n)$ lớn hơn đáng kể $\implies$ Trường hợp 3.
* **Độ phức tạp:** $T(n) = \mathbf{\Theta(n^2)}$.
* **Kiến thức liên quan:** Định lý Thợ, Phân tích thuật toán đệ quy.

**Master Theorem:** Định lý áp dụng cho các hệ thức truy hồi có dạng:

  $$T(n) = aT\left(\frac{n}{b}\right) + f(n)$$
  
  Trong đó:

* $n$: Kích thước của bài toán ban đầu.
* $a$: Số lượng bài toán con trong mỗi bước đệ quy ($a \geq 1$).
* $n/b$: Kích thước của mỗi bài toán con ($b > 1$).
* $f(n)$: Chi phí của các thao tác ngoài đệ quy (như chia bài toán và hợp nhất kết quả).

### 19. Bài toán: Độ phức tạp tìm kiếm chuỗi trong Trie

* **Yêu cầu:** Tìm kiếm chuỗi $S$ có độ dài $d$ trong Trie.
* **Độ phức tạp:** Bằng độ dài của chuỗi, $\mathbf{O(d)}$.
* **Kết quả (Trắc nghiệm):** $\mathbf{O(n)}$ (Giả định $n$ là độ dài chuỗi $S$ trong bối cảnh này).
* **Kiến thức liên quan:** Cấu trúc dữ liệu Trie, Độ phức tạp thời gian.
